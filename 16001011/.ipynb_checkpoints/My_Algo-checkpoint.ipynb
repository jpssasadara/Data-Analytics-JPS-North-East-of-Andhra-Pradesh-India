```python
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Trainning dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

##################### Categorical Features ################################
from keras.utils import np_utils
#Get gender coloum dataFrame format
#df=df["Gender"]  
#Get whole data frame and replace categorical fields with numeric values.
df=df.replace(to_replace ='Female', value =0) 
df=df.replace(to_replace ='Male', value =1) 
#print(df)

value = df.values  # change data frame to values
Genderdata=value[:len(data),2]

print(Genderdata)

#Convert to Categorical Reprecentation 
Categorical_gender = np_utils.to_categorical(Genderdata)
print(Categorical_gender[:10])
###########################################################################

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Testing dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

#####################  Categorical Features ###########################
from keras.utils import np_utils
#Get gender coloum dataFrame format
#df=df["Gender"]  
#Get whole data frame and replace categorical fields with numeric values.
dfTest=dfTest.replace(to_replace ='Female', value =0) 
dfTest=dfTest.replace(to_replace ='Male', value =1) 
#print(df)

valueTest = dfTest.values  # change data frame to values
GenderdataTest = valueTest[:len(dataTest),2]

print(GenderdataTest)

#Convert to Categorical Reprecentation 
Categorical_genderTest = np_utils.to_categorical(GenderdataTest)
print(Categorical_genderTest[:10])
#########################################################################

#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
```

```python
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Trainning dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

##################### Creating final dataset #######################
FinalDatasetLable = []
FinalDatasetTarget = []
for x in range(len(data)):
    FinalDatasetTarget.append(dataset[x][11])
                                             #convert numpy array to list
    FinalDatasetLable.append([dataset[x][1]]+Fixedpartdata[x].tolist()+Categorical_gender[x].tolist())
    
    #.append(Fixedpartdata[x])
    #.append(Categorical_gender[x])
    
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Testing dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$

FinalDatasetLableTest = []
FinalDatasetTargetTest = []
for x in range(len(dataTest)):
    FinalDatasetTargetTest.append(datasetTest[x][11])
                                             #convert numpy array to list
    FinalDatasetLableTest.append([datasetTest[x][1]]+FixedpartdataTest[x].tolist()+Categorical_genderTest[x].tolist())
    
    #.append(Fixedpartdata[x])
    #.append(Categorical_gender[x])
    
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
```

```python
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Trainning dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#Convert list to numpy array--> for Tenserflow input
FinalDatasetLable = numpy.array(FinalDatasetLable)
FinalDatasetTarget = numpy.array(FinalDatasetTarget)
print(FinalDatasetTarget[:5])
print(FinalDatasetLable[:10])
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$ Testing dataset $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
#Convert list to numpy array--> for Tenserflow input
FinalDatasetLableTest = numpy.array(FinalDatasetLableTest)
FinalDatasetTargetTest = numpy.array(FinalDatasetTargetTest)
print(FinalDatasetTargetTest[:5])
print(FinalDatasetLableTest[:10])
#$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$
```

```python
from keras.layers import Dense,Dropout #densly connected layers
from keras.models import Sequential


modelTrain=Sequential() #model is an empty NN
#1st hidden layer (dense type-fully connected)
modelTrain.add(Dense(44,input_dim=11,activation='relu'))
#2nd Hidden layer
modelTrain.add(Dense(22,input_dim=44,activation='relu'))
#model.add(Dropout(0.5))
#3rd Hidden layer
modelTrain.add(Dense(11,input_dim=22,activation='relu'))
#4rd Hidden layer
modelTrain.add(Dense(5,input_dim=11,activation='relu'))
#final layer
modelTrain.add(Dense(2,input_dim=5,activation='softmax'))

modelTrain.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```
